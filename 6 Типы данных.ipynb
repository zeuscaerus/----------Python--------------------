–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:

–î–∞–Ω–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ. –í—ã–≤–µ–¥–∏—Ç–µ –µ–≥–æ –¥—Ä–æ–±–Ω—É—é —á–∞—Å—Ç—å.

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –¥—Ä–æ–±–Ω—É—é —á–∞—Å—Ç—å —á–∏—Å–ª–∞ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

–ù–∞–∏–±–æ–ª—å—à–µ–µ –∏ –Ω–∞–∏–º–µ–Ω—å—à–µ–µ

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –∏ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∏–∑ –ø—è—Ç–∏ —á–∏—Å–µ–ª.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –ø—è—Ç—å —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –∏ –Ω–∞–∏–±–æ–ª—å—à–µ–µ —á–∏—Å–ª–æ —Å –ø–æ—è—Å–Ω—è—é—â–µ–π –Ω–∞–¥–ø–∏—Å—å—é.

```python
a=(int(input()),int(input()),int(input()),int(input()),int(input()))
b=min(a)
c=max(a)
print(f'–ù–∞–∏–º–µ–Ω—å—à–µ–µ —á–∏—Å–ª–æ = {b}\n–ù–∞–∏–±–æ–ª—å—à–µ–µ —á–∏—Å–ª–æ = {c}')
```

–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ —Ç—Ä—ë—Ö üå∂Ô∏è

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —É–ø–æ—Ä—è–¥–æ—á–∏–≤–∞–µ—Ç —Ç—Ä–∏ —á–∏—Å–ª–∞ –æ—Ç –±–æ–ª—å—à–µ–≥–æ –∫ –º–µ–Ω—å—à–µ–º—É.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Ç—Ä–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç—Ä–∏ —á–∏—Å–ª–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ, —É–ø–æ—Ä—è–¥–æ—á–µ–Ω–Ω—ã—Ö –æ—Ç –±–æ–ª—å—à–µ–≥–æ –∫ –º–µ–Ω—å—à–µ–º—É.

```python
a,b,c=int(input()), int(input()), int(input())
x=(a,b,c)
xmin=min(x)
xmax=max(x)
xmid=(a+b+c)-(xmin+xmax)
print(xmax,xmid,xmin, sep='\n')
```

–ò–Ω—Ç–µ—Ä–µ—Å–Ω–æ–µ —á–∏—Å–ª–æ

–ù–∞–∑–æ–≤–µ–º —á–∏—Å–ª–æ –∏–Ω—Ç–µ—Ä–µ—Å–Ω—ã–º, –µ—Å–ª–∏ –≤ –Ω–µ–º —Ä–∞–∑–Ω–æ—Å—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –∏ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π —Ü–∏—Ñ—Ä—ã —Ä–∞–≤–Ω—è–µ—Ç—Å—è —Å—Ä–µ–¥–Ω–µ–π –ø–æ –≤–µ–ª–∏—á–∏–Ω–µ —Ü–∏—Ñ—Ä–µ. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ–µ —á–∏—Å–ª–æ –∏–ª–∏ –Ω–µ—Ç. –ï—Å–ª–∏ —á–∏—Å–ª–æ –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ–µ, —Å–ª–µ–¥—É–µ—Ç –≤—ã–≤–µ—Å—Ç–∏ ‚Äì ¬´–ß–∏—Å–ª–æ –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ–µ¬ª –∏–Ω–∞—á–µ ¬´–ß–∏—Å–ª–æ –Ω–µ–∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ–µ¬ª.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Ü–µ–ª–æ–µ —Ç—Ä–µ—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

```python
a,b,c=input()
a=int(a)
b=int(b)
c=int(c)
x=(a,b,c)
xmin=min(x)
xmax=max(x)
xmid=(a+b+c)-(xmin+xmax)
if (xmax-xmin)==xmid:
    print('–ß–∏—Å–ª–æ –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ–µ')
else:
    print('–ß–∏—Å–ª–æ –Ω–µ–∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ–µ')
```

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äì –º–∞–Ω—Ö—ç—Ç—Ç–µ–Ω—Å–∫–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ.

–¢—Ä–∏ –≥–æ—Ä–æ–¥–∞

–î–∞–Ω—ã –Ω–∞–∑–≤–∞–Ω–∏—è —Ç—Ä–µ—Ö –≥–æ—Ä–æ–¥–æ–≤. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç —Å–∞–º–æ–µ –∫–æ—Ä–æ—Ç–∫–æ–µ –∏ —Å–∞–º–æ–µ –¥–ª–∏–Ω–Ω–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –≥–æ—Ä–æ–¥–∞.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –Ω–∞–∑–≤–∞–Ω–∏—è —Ç—Ä–µ—Ö –≥–æ—Ä–æ–¥–æ–≤, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Å–∞–º–æ–µ –∫–æ—Ä–æ—Ç–∫–æ–µ –∏ –¥–ª–∏–Ω–Ω–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –≥–æ—Ä–æ–¥–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ –¥–ª–∏–Ω—ã –Ω–∞–∑–≤–∞–Ω–∏–π –≤—Å–µ—Ö —Ç—Ä–µ—Ö –≥–æ—Ä–æ–¥–æ–≤ —Ä–∞–∑–ª–∏—á–Ω—ã.

```python
t1,t2,t3=input(), input(), input()
a=len(t1)
b=len(t2)
c=len(t3)
a=int(a)
b=int(b)
c=int(c)
x=(a,b,c)
xmin=min(x)
xmax=max(x)
if xmin==a:
    print(t1)
elif xmin==b:
    print(t2)
else:
    print(t3)
if xmax==a:
    print(t1)
elif xmax==b:
    print(t2)
else:
    print(t3)
```

–ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏–µ —Å—Ç—Ä–æ–∫–∏

–í–≤–æ–¥—è—Ç—Å—è 3 —Å—Ç—Ä–æ–∫–∏ –≤ —Å–ª—É—á–∞–π–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—è—Å–Ω—è–µ—Ç –º–æ–∂–Ω–æ –ª–∏ –∏–∑ –¥–ª–∏–Ω —ç—Ç–∏—Ö —Å—Ç—Ä–æ–∫ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≤–æ–∑—Ä–∞—Å—Ç–∞—é—â—É—é –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫—É—é –ø—Ä–æ–≥—Ä–µ—Å—Å–∏—é.

```python
a,b,c=input(),input(),input()
a1=len(a)
b1=len(b)
c1=len(c)
x=(a1,b1,c1)
xmin=min(x)
xmax=max(x)
xmid=(a1+b1+c1)-(xmax+xmin)
if (xmax-xmid)==(xmid-xmin):
    print('YES')
else:
    print('NO')
```

–¶–≤–µ—Ç –Ω–∞—Å—Ç—Ä–æ–µ–Ω–∏—è —Å–∏–Ω–∏–π

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç—ã–≤–∞–µ—Ç –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É, –ø–æ—Å–ª–µ —á–µ–≥–æ –≤—ã–≤–æ–¥–∏—Ç ¬´YES¬ª, –µ—Å–ª–∏ –≤ –≤–≤–µ–¥–µ–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –µ—Å—Ç—å –ø–æ–¥—Å—Ç—Ä–æ–∫–∞ ¬´—Å–∏–Ω–∏–π¬ª –∏ ¬´NO¬ª –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

–û—Ç–¥—ã—Ö–∞–µ–º –ª–∏?

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç—ã–≤–∞–µ—Ç –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É, –ø–æ—Å–ª–µ —á–µ–≥–æ –≤—ã–≤–æ–¥–∏—Ç ¬´YES¬ª, –µ—Å–ª–∏ –≤ –≤–≤–µ–¥—ë–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –µ—Å—Ç—å –ø–æ–¥—Å—Ç—Ä–æ–∫–∞ ¬´—Å—É–±–±–æ—Ç–∞¬ª –∏–ª–∏ ¬´–≤–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ¬ª, –∏ ¬´NO¬ª –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

```python
x=input()
sd='—Å—É–±–±–æ—Ç–∞'
sund='–≤–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ'
if sd in x or sund in x:
    print('YES')
else:
    print('NO')
```

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –æ–¥–Ω–∞ —Å—Ç—Ä–æ–∫–∞ ‚Äì email –∞–¥—Ä–µ—Å.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Å—Ç—Ä–æ–∫—É ¬´YES¬ª, –µ—Å–ª–∏ email –∞–¥—Ä–µ—Å —è–≤–ª—è–µ—Ç—Å—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–º –∏ ¬´NO¬ª –≤ –∏–Ω–æ–º —Å–ª—É—á–∞–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ù–∞–ª–∏—á–∏–µ —Å–∏–º–≤–æ–ª–æ–≤ @ –∏ . –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ email –∞–¥—Ä–µ—Å–∞, –æ–¥–Ω–∞–∫–æ –∏—Ö –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–µ –≥–∞—Ä–∞–Ω—Ç–∏—Ä–æ–≤–∞–Ω–æ –≤–ª–µ—á—ë—Ç –∑–∞ —Å–æ–±–æ–π –Ω–µ–≤–µ—Ä–Ω—ã–π email.

```python
x=input()
sd='@'
sund='.'
if sd in x and sund in x:
    print('YES')
else:
    print('NO')
```

6.3 –ú–æ–¥—É–ª—å math

–ï–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ

–ù–∞ –ø–ª–æ—Å–∫–æ—Å—Ç–∏ –µ–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –¥–≤—É–º—è —Ç–æ—á–∫–∞–º–∏ (x1; y1) –∏ (x2; y2) –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è —Ç–∞–∫ œÅ= (–∫–æ—Ä–µ–Ω—å(x1‚àíx2)^2+(y1‚àíy2)^2).

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –æ–ø—Ä–µ–¥–µ–ª—è—é—â—É—é –µ–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –¥–≤—É–º—è —Ç–æ—á–∫–∞–º–∏, –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –∫–æ—Ç–æ—Ä—ã—Ö –∑–∞–¥–∞–Ω—ã.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —á–µ—Ç—ã—Ä–µ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–ª–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ ‚Äì x1, y1, x2, y2‚Äã.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äì –µ–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ.

```python
from math import sqrt
x1,y1,x2,y2=float(input()),float(input()),float(input()),float(input())
p=sqrt((x1-x2)**2+(y1-y2)**2)
print(p)
```

–ü–ª–æ—â–∞–¥—å –∏ –¥–ª–∏–Ω–∞

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –æ–ø—Ä–µ–¥–µ–ª—è—é—â—É—é –ø–ª–æ—â–∞–¥—å –∫—Ä—É–≥–∞ –∏ –¥–ª–∏–Ω—É –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏ –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É —Ä–∞–¥–∏—É—Å—É R.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ R‚Äã.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –¥–≤–∞ —á–∏—Å–ª–∞ ‚Äì –ø–ª–æ—â–∞–¥—å –∫—Ä—É–≥–∞ –∏ –¥–ª–∏–Ω—É –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏ —Ä–∞–¥–∏—É—Å–∞ R.

–ü—Ä–∏–º–∏—á–∞–Ω–∏–µ. S=pi * R^2, C=2 * pi * R. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–æ–Ω—Å—Ç–∞–Ω—Ç—É math.pi.

```python
from math import pi
r=float(input())
print(pi*r**2)
print(2*pi*r)
```

–°—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è
–í –º–∞—Ç–µ–º–∞—Ç–∏–∫–µ –≤—ã–¥–µ–ª—è—é—Ç —Å–ª–µ–¥—É—é—â–∏–µ —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è:

—Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —á–∏—Å–µ–ª a –∏ b: (a + b)/2 ;</br>
—Å—Ä–µ–¥–Ω–µ–µ –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–µ —á–∏—Å–µ–ª a –∏ b: –∫–æ—Ä–µ–Ω—å(a * b);</br>
—Å—Ä–µ–¥–Ω–µ–µ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ —á–∏—Å–µ–ª a –∏ b: (2 * a * b)/(a + b);</br>
 —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ —á–∏—Å–µ–ª a –∏ b: –∫–æ—Ä–µ–Ω—å((a^2 + b^2)/2).

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –¥–≤–∞ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–ª–∞ a –∏ b‚Äã, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ 4 —á–∏—Å–ª–∞ ‚Äì —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ, –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–µ, –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ –∏ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ.

```python
from math import *
a, b=float(input()), float(input())
print((a+b)/2)
print(sqrt(a*b))
print((2*a*b)/(a+b))
print(sqrt((a**2+b**2)/2))
```

–¢—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ 

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –≤—ã—á–∏—Å–ª—è—é—â—É—é –∑–Ω–∞—á–µ–Ω–∏–µ —Ç—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è sinx+cosx+tan^2x –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É —á–∏—Å–ª—É –≥—Ä–∞–¥—É—Å–æ–≤ x.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ x –∏–∑–º–µ—Ä—è–µ–º–æ–µ –≤ –≥—Ä–∞–¥—É—Å–∞—Ö‚Äã. 

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äì –∑–Ω–∞—á–µ–Ω–∏–µ —Ç—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 1. –¢—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –ø—Ä–∏–Ω–∏–º–∞—é—Ç –∞—Ä–≥—É–º–µ–Ω—Ç –≤ —Ä–∞–¥–∏–∞–Ω–∞—Ö. –ß—Ç–æ–±—ã –ø–µ—Ä–µ–≤–µ—Å—Ç–∏ –≥—Ä–∞–¥—É—Å—ã –≤ —Ä–∞–¥–∏–∞–Ω—ã, –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å —Ñ–æ—Ä–º—É–ª–æ–π r=(x * pi)/180.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. –ú–æ–¥—É–ª—å math —Å–æ–¥–µ—Ä–∂–∏—Ç –≤—Å—Ç—Ä–æ–µ–Ω–Ω—É—é —Ñ—É–Ω–∫—Ü–∏—é radians(), –∫–æ—Ç–æ—Ä–∞—è –ø–µ—Ä–µ–≤–æ–¥–∏—Ç —É–≥–æ–ª –∏–∑ –≥—Ä–∞–¥—É—Å–æ–≤ –≤ —É–≥–æ–ª –≤ —Ä–∞–¥–∏–∞–Ω–∞—Ö.

```python
from math import *
x=float(input())
r=(x*pi)/180
print(sin(r)+cos(r)+tan(r)**2)
```

–ü–æ–ª –∏ –ø–æ—Ç–æ–ª–æ–∫

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –≤—ã—á–∏—Å–ª—è—é—â—É—é –∑–Ω–∞—á–µ–Ω–∏–µ ‚åàx‚åâ +‚åäx‚åã –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–º—É —á–∏—Å–ª—É x.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ x.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äì –∑–Ω–∞—á–µ–Ω–∏–µ —É–∫–∞–∑–∞–Ω–Ω–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. ‚åàx‚åâ ‚Äì –ø–æ—Ç–æ–ª–æ–∫ —á–∏—Å–ª–∞, ‚åäx‚åã ‚Äì –ø–æ–ª —á–∏—Å–ª–∞.

```python
from math import ceil, floor
x=float(input())
x1=ceil(x)
x2=floor(x)
print(x1+x2)
```

–ö–≤–∞–¥—Ä–∞—Ç–Ω–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ üå∂Ô∏èüå∂Ô∏è

–î–∞–Ω—ã —Ç—Ä–∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–ª–∞ a, b, c. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –∫–æ—Ä–Ω–∏ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ —É—Ä–∞–≤–Ω–µ–Ω–∏—è ax^2+bx+c=0.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Ç—Ä–∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–ª–∞ a ‚â† 0, b, c, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –∫–æ—Ä–Ω–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –µ—Å–ª–∏ –æ–Ω–∏ —Å—É—â–µ—Å—Ç–≤—É—é—Ç –∏–ª–∏ —Ç–µ–∫—Å—Ç ¬´–ù–µ—Ç –∫–æ—Ä–Ω–µ–π¬ª –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ï—Å–ª–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ –∏–º–µ–µ—Ç –¥–≤–∞ –∫–æ—Ä–Ω—è, —Ç–æ —Å–ª–µ–¥—É–µ—Ç –≤—ã–≤–µ—Å—Ç–∏ –∏—Ö –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è.

```python
a,b,c=float(input()), float(input()), float(input())
d=b**2-4*a*c
if d<0:
    print('–ù–µ—Ç –∫–æ—Ä–Ω–µ–π')
elif d==0:
    print(-b/(2*a))
elif d>0:
    x1=(-b+d**0.5)/(2*a)
    x2=(-b-d**0.5)/(2*a)
    print(min(x1,x2), max(x1,x2), sep='\n')
```

–ü—Ä–∞–≤–∏–ª—å–Ω—ã–π –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫

–ü—Ä–∞–≤–∏–ª—å–Ω—ã–π –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫ ‚Äî –≤—ã–ø—É–∫–ª—ã–π –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫, —É –∫–æ—Ç–æ—Ä–æ–≥–æ —Ä–∞–≤–Ω—ã –≤—Å–µ —Å—Ç–æ—Ä–æ–Ω—ã –∏ –≤—Å–µ —É–≥–ª—ã –º–µ–∂–¥—É —Å–º–µ–∂–Ω—ã–º–∏ —Å—Ç–æ—Ä–æ–Ω–∞–º–∏. –ü–ª–æ—â–∞–¥—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ —Å –¥–ª–∏–Ω–æ–π —Å—Ç–æ—Ä–æ–Ω—ã a –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —Å—Ç–æ—Ä–æ–Ω n –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ: S = (n * a^2)/4 * tg * (pi/n).

–î–∞–Ω—ã –¥–≤–∞ —á–∏—Å–ª–∞: –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n –∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ a. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞—Ö–æ–¥–∏—Ç –ø–ª–æ—â–∞–¥—å —É–∫–∞–∑–∞–Ω–Ω–æ–≥–æ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –¥–≤–∞ —á–∏—Å–ª–∞ n –∏ a, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ ‚Äì –ø–ª–æ—â–∞–¥—å –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞.

```python
from math import*
n,a=float(input()), float(input())
s=(n*a**2)/(4*tan(pi/n))
print(s)
```
