–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ —Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–∞–π—Ç–∏ –æ—à–∏–±–∫–∏ –≤ –∏–º–µ—é—â–µ–π—Å—è –ø—Ä–æ–≥—Ä–∞–º–º–µ, –∞ –Ω–µ –Ω–∞–ø–∏—Å–∞—Ç—å —Å–≤–æ—é, –≤–æ–∑–º–æ–∂–Ω–æ, –∏—Å–ø–æ–ª—å–∑—É—é—â—É—é –¥—Ä—É–≥–æ–π –∞–ª–≥–æ—Ä–∏—Ç–º —Ä–µ—à–µ–Ω–∏—è.

```python
count = 0
maximum = -10**6
for i in range(8):
    x = int(input())
    if x % 4 == 0:
        count += 1
        if x > maximum:
            maximum = x
if count > 0:
    print(count)
    print(maximum)
else:
    print('NO')
```

–ù–∞ –æ–±—Ä–∞–±–æ—Ç–∫—É –ø–æ—Å—Ç—É–ø–∞–µ—Ç –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∏–∑ 4 —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª. –ò–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –≤–≤–æ–¥–∏–º—ã–µ —á–∏—Å–ª–∞ –ø–æ –∞–±—Å–æ–ª—é—Ç–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω–µ –Ω–µ –ø—Ä–µ–≤—ã—à–∞—é—Ç 10^8. –ù—É–∂–Ω–æ –Ω–∞–ø–∏—Å–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –Ω–∞ —ç–∫—Ä–∞–Ω –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ—á—ë—Ç–Ω—ã—Ö —á–∏—Å–µ–ª –≤ –∏—Å—Ö–æ–¥–Ω–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –Ω–µ—á—ë—Ç–Ω–æ–µ —á–∏—Å–ª–æ. –ï—Å–ª–∏ –Ω–µ—á—ë—Ç–Ω—ã—Ö —á–∏—Å–µ–ª –Ω–µ—Ç, —Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–∞ —ç–∫—Ä–∞–Ω –≤—ã–≤–µ—Å—Ç–∏ ¬´NO¬ª. –ü—Ä–æ–≥—Ä–∞–º–º–∏—Å—Ç —Ç–æ—Ä–æ–ø–∏–ª—Å—è –∏ –Ω–∞–ø–∏—Å–∞–ª –ø—Ä–æ–≥—Ä–∞–º–º—É –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ.

–ù–∞–π–¥–∏—Ç–µ –≤—Å–µ –æ—à–∏–±–∫–∏ –≤ —ç—Ç–æ–π –ø—Ä–æ–≥—Ä–∞–º–º–µ (–∏—Ö –º–æ–∂–µ—Ç –±—ã—Ç—å –æ–¥–Ω–∞ –∏–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–æ). –ò–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –∫–∞–∂–¥–∞—è –æ—à–∏–±–∫–∞ –∑–∞—Ç—Ä–∞–≥–∏–≤–∞–µ—Ç —Ç–æ–ª—å–∫–æ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É –∏ –º–æ–∂–µ—Ç –±—ã—Ç—å –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–∞ –±–µ–∑ –∏–∑–º–µ–Ω–µ–Ω–∏—è –¥—Ä—É–≥–∏—Ö —Å—Ç—Ä–æ–∫.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ —Ç—Ä–µ–±—É–µ—Ç—Å—è –Ω–∞–π—Ç–∏ –æ—à–∏–±–∫–∏ –≤ –∏–º–µ—é—â–µ–π—Å—è –ø—Ä–æ–≥—Ä–∞–º–º–µ, –∞ –Ω–µ –Ω–∞–ø–∏—Å–∞—Ç—å —Å–≤–æ—é, –≤–æ–∑–º–æ–∂–Ω–æ, –∏—Å–ø–æ–ª—å–∑—É—é—â—É—é –¥—Ä—É–≥–æ–π –∞–ª–≥–æ—Ä–∏—Ç–º —Ä–µ—à–µ–Ω–∏—è.

```python
count = 0
maximum = -10**6
for i in range(4):
    x = int(input())
    if x % 2 != 0:
        count += 1
        if x > maximum:
            maximum = x
if count > 0:
    print(count)
    print(maximum)
else:
    print('NO')
```

–ó–≤–µ–∑–¥–Ω–∞—è —Ä–∞–º–∫–∞

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–µ—á–∞—Ç–∞–µ—Ç –∑–≤–µ–∑–¥–Ω—É—é —Ä–∞–º–∫—É —Ä–∞–∑–º–µ—Ä–∞–º–∏ n√ó19.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n‚àà[3; 19] ‚Äî –≤—ã—Å–æ—Ç–∞ –∑–≤–µ–∑–¥–Ω–æ–π —Ä–∞–º–∫–∏.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –∑–≤–µ–∑–¥–Ω—É—é —Ä–∞–º–∫—É —Ä–∞–∑–º–µ—Ä–∞–º–∏ n√ó19.

–ü–æ–¥—Å–∫–∞–∑–∫–∞. –î–ª—è –ø–µ—á–∞—Ç–∏ –∑–≤–µ–∑–¥–Ω–æ–π –ª–∏–Ω–∏–∏ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —É–º–Ω–æ–∂–µ–Ω–∏–µ —Å—Ç—Ä–æ–∫–∏ –Ω–∞ —á–∏—Å–ª–æ.

```python
x=int(input())
for i in range (1,x+1):
    if i==1 or i==x:
        print('*'*19)
    else:
        print('*                 *')
```

–¢—Ä–µ—Ç—å—è —Ü–∏—Ñ—Ä–∞

–î–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ n(n>99). –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –µ–≥–æ —Ç—Ä–µ—Ç—å—é (—Å –Ω–∞—á–∞–ª–∞) —Ü–∏—Ñ—Ä—É.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ, —Å–æ—Å—Ç–æ—è—â–µ–µ –∫–∞–∫ –º–∏–Ω–∏–º—É–º –∏–∑ —Ç—Ä–µ—Ö —Ü–∏—Ñ—Ä.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –µ–≥–æ —Ç—Ä–µ—Ç—å—é (—Å –Ω–∞—á–∞–ª–∞) —Ü–∏—Ñ—Ä—É.

```python
x=int(input())
num=0
while x>99:
    y=x%10
    num=y
    x//=10
print(num)
```

–í—Å–µ –≤–º–µ—Å—Ç–µ 2

–î–∞–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã—á–∏—Å–ª—è–µ—Ç:

–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ü–∏—Ñ—Ä 3 –≤ –Ω–µ–º;</br>
—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –≤ –Ω–µ–º –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –ø–æ—Å–ª–µ–¥–Ω—è—è —Ü–∏—Ñ—Ä–∞;</br>
–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–µ—Ç–Ω—ã—Ö —Ü–∏—Ñ—Ä;</br>
—Å—É–º–º—É –µ–≥–æ —Ü–∏—Ñ—Ä, –±–æ–ª—å—à–∏—Ö –ø—è—Ç–∏;</br>
–ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ü–∏—Ñ—Ä, –±–æ–ª—å—à–∏—Ö —Å–µ–º–∏ (–µ—Å–ª–∏ —Ü–∏—Ñ—Ä –±–æ–ª—å—à–∏—Ö —Å–µ–º–∏ –Ω–µ—Ç, —Ç–æ –≤—ã–≤–µ—Å—Ç–∏ 1, –µ—Å–ª–∏ —Ç–∞–∫–∞—è —Ü–∏—Ñ—Ä–∞ –æ–¥–Ω–∞, —Ç–æ –≤—ã–≤–µ—Å—Ç–∏ –µ–µ);</br>
—Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –≤ –Ω–µ–º –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è —Ü–∏—Ñ—Ä—ã 0 –∏ 5 (–≤—Å–µ–≥–æ —Å—É–º–º–∞—Ä–Ω–æ).

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏—è —É–∫–∞–∑–∞–Ω–Ω—ã—Ö –≤–µ–ª–∏—á–∏–Ω –≤ —É–∫–∞–∑–∞–Ω–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ.

```python
x=int(input())
num_3=0
num_last=x%10
num_last_count=0
num_chet=0
num_more_five=0
num_more_seven=1
zero_five=0
while x>0:
    y=x%10
    if y==3:
        num_3+=1
    if y==num_last:
        num_last_count+=1
    if y%2==0:
        num_chet+=1
    if y>5:
        num_more_five+=y
    if y>7:
        num_more_seven*=y
    if y==0 or y==5:
        zero_five+=1
    x//=10
print(num_3)
print(num_last_count)
print(num_chet)
print(num_more_five)
print(num_more_seven)
print(zero_five)
```
