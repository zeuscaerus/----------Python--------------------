15.2 Числовая угадайка

Угадайка чисел

Описание проекта: программа генерирует случайное число в диапазоне от 1 до 100 и просит пользователя угадать это число. Если догадка пользователя больше случайного числа, то программа должна вывести сообщение 'Слишком много, попробуйте еще раз'. Если догадка меньше случайного числа, то программа должна вывести сообщение 'Слишком мало, попробуйте еще раз'. Если пользователь угадывает число, то программа должна поздравить его и вывести сообщение 'Вы угадали, поздравляем!'.

```python
import random
def is_valid(num):
    return num.isdigit() and 0<int(num)<101
def is_valid_num(num):
    return num.isdigit() and int(num)>1


x2=input('Укажите предельную границу числа для угадывания: ')
while is_valid_num(x2) is False:
    x2=input('Укажите число: ')
else:
    x=random.randint(1,int(x2))
count=0     
again='д'
print('Добро пожаловать в числовую угадайку')
while again=='д':
    number=input(f'Введите число от 1 до {x2}: ')
    while is_valid(number) is False:
        count+=1
        print(f'А может быть все-таки введем целое число от 1 до {x2}?')
        number=input(f'Введите число от 1 до {x2}: ')
    while is_valid(number) is True:
        count+=1
        if int(number)==x:
            print(f'Вы угадали число {x} за {count} хода(ов), поздравляем!\nСпасибо, что играли в числовую угадайку. Еще увидимся...')
            again=input('Хотите сыграть еще раз? да(д)/нет(н): ') 
            break
        elif int(number)<x:
            print(f'Ваше число {number} меньше загаданного на {x-int(number)}, попробуйте еще раз')
            number=input(f'Введите число от 1 до {x2}: ') 
        elif int(number)>x:
            print(f'Ваше число {number} больше загаданного на {int(number)-x}, попробуйте еще раз')
            number=input(f'Введите число от 1 до {x2}: ') 
```

На вход программе подается натуральное число n.

Формат выходных данных: 
Программа должна вывести наименьшее количество вопросов, которых гарантированно хватит Руслану, чтобы угадать число Тимура.

15.4 Генератор безопасных паролей

Генератор безопасных паролей

Описание проекта: программа генерирует заданное количество паролей и включает в себя умную настройку на длину пароля, а также на то, какие символы требуется в него включить, а какие исключить.

```python
import random
import string

lower=string.ascii_lowercase
upper=string.ascii_uppercase
num=string.digits
symbol=string.punctuation
symbol2=['i', 'l', '1', 'L', 'o', '0', 'O']

print('Привет, я генератор паролей!\n')
again='д'
while again=='д':

    quantity=int(input('\nВведите количество паролей:'))
    length=int(input('\nВведите длинну пароля: '))
    nums=input('Нужны ли цифры в пароле? да/нет: ').lower()
    lettersdown=input('Нужны ли строчные буквы? да/нет: ').lower()
    lettersup=input('Нужны ли заглавные буквы? да/нет: ').lower()
    symbols=input('Нужны ли символы? да/нет: ').lower()
    symbols2=input('Нужно ли исключить неоднозначные символы "il1Lo0O"? да/нет: ').lower()

    all2=[]


    for i in range(quantity):
        if nums=='д':
            all2+=num
        else:
            continue
        if lettersdown=='д':
            all2+=lower
        else:
            continue
        if lettersup=='д':
            all2+=upper
        else:
            continue
        if symbols=='д':
            all2+=symbol
        else:
            continue
        if symbols2=='н':
            all2+=symbol2
        else:
            continue
        temp=random.sample(all2,length)
        password=''.join(temp)
        print(password)
    again=input('Продолжить? да/нет: ')
```

15.5 Шифр Цезаря

Шифр Цезаря

Шифр Цезаря (шифр сдвига) — один из самых простых и наиболее широко известных методов шифрования. Шифр Цезаря — это вид шифра подстановки, в котором каждый символ в открытом тексте заменяется символом, находящимся на некотором постоянном числе позиций левее или правее него в алфавите.

```python
def crypto_ru(text, num):
    temp=[ord(j) for j in text]
    temp2=[]
    for i in range(len(temp)):
        if temp[i]<65:
            temp2.append(temp[i])
        elif 1039<temp[i]<1072:
            let=temp[i]+num
            if let>1072:
                temp2.append(1039+(let-1071))
            else:
                temp2.append(let)
        elif 1071<temp[i]<1104:
            let=temp[i]+num
            if let>1103:
                temp2.append(1071+(let-1103))
            else:
                temp2.append(let)
    shifr=[chr(j) for j in temp2]
    return ''.join(shifr)

def crypto_eng(text, num):
    temp=[ord(j) for j in text]
    temp2=[]
    for i in range(len(temp)):
        if temp[i]<65:
            temp2.append(temp[i])
        elif 64<temp[i]<91:
            let=temp[i]+num
            if let>90:
                temp2.append(64+(let-90))
            else:
                temp2.append(let)
        elif 96<temp[i]<123:
            let=temp[i]+num
            if let>122:
                temp2.append(96+(let-122))
            else:
                temp2.append(let)
    shifr=[chr(j) for j in temp2]
    return ''.join(shifr)

def de_crypto_ru(text, num):
    temp=[ord(j) for j in text]
    temp2=[]
    for i in range(len(temp)):
        if temp[i]<65:
            temp2.append(temp[i])
        elif 1039<temp[i]<1072:
            let=temp[i]-num
            if let<1040:
                temp2.append(1072-(1040-let))
            else:
                temp2.append(let)
        elif 1071<temp[i]<1104:
            let=temp[i]-num
            if let<1072:
                temp2.append(1104-(1072-let))
            else:
                temp2.append(let)
    shifr=[chr(j) for j in temp2]
    return ''.join(shifr)

def de_crypto_eng(text, num):
    temp=[ord(j) for j in text]
    temp2=[]
    for i in range(len(temp)):
        if temp[i]<65:
            temp2.append(temp[i])
        elif 64<temp[i]<91:
            let=temp[i]-num
            if let<65:
                temp2.append(91-(65-let))
            else:
                temp2.append(let)
        elif 96<temp[i]<123:
            let=temp[i]-num
            if let<97:
                temp2.append(123-(97-let))
            else:
                temp2.append(let)
    shifr=[chr(j) for j in temp2]
    return ''.join(shifr)

print('Добропожаловать в программу по зашифровке и расшифровки по принципу Цезаря!')
name=input('Введите ваше имя: ')
again='д'
while again=='д':
    action=input(f'{name}, Вы хотите зашифровать или расшифровать текст? (з/р): ')
    lang=input('На каком языке будет текст? (рус/англ): ').lower()
    rot=int(input('Введите значение сдвига: '))
    text=input('Введите текст: ')
    if action=='з' and lang=='рус':
        print(crypto_ru(text, rot))
        again=input('Хотите ввести еще один текст для зашифровки/расшифровки на русском/английском языках? (д/н): ')
    if action=='з' and lang=='англ':
        print(crypto_eng(text, rot))
        again=input('Хотите ввести еще один текст для зашифровки/расшифровки на русском/английском языках? (д/н): ')
    if action=='р' and lang=='рус':
        print(de_crypto_ru(text, rot))
        again=input('Хотите ввести еще один текст для зашифровки/расшифровки на русском/английском языках? (д/н): ')
    if action=='р' and lang=='англ':
        print(de_crypto_eng(text,rot))
        again=input('Хотите ввести еще один текст для зашифровки/расшифровки на русском/английском языках? (д/н): ')
```

BOH

На вход программе подается натуральное число в десятичной системе счисления. Напишите программу, которая переводит его в двоичную, восьмеричную и шестнадцатеричную системы счисления.

Формат входных данных: 
На вход программе подается натуральное число.

Формат выходных данных: 
Программа должна вывести текст в соответствии с условием задачи.

Примечание 1. Используйте встроенные функции bin(), oct(), hex().

Примечание 2. Для шестнадцатеричной системы счисления используйте заглавные буквы A, B, C, D, E, F.

Примечание 3. BOH = Binary, Octal, Hex.

```python
def bin_oct_hex(num):
    bin_num=bin(num)
    oct_num=oct(num)
    hex_num=hex(num)
    return print(bin_num[2:], oct_num[2:], hex_num[2:].upper(), sep='\n')

x=int(input())

print(bin_oct_hex(x))
```

15.7 Угадайка слов

Угадайка слов

Описание проекта: программа загадывает слово, а пользователь должен его угадать. Изначально все буквы слова неизвестны. Также рисуется виселица с петлей. Пользователь предлагает букву, которая может входить в это слово. Если такая буква есть в слове, то программа ставит букву столько раз, сколько она встречается в слове. Если такой буквы нет, к виселице добавляется круг в петле, изображающий голову. Пользователь продолжает отгадывать буквы до тех пор, пока не отгадает всё слово. За каждую неудачную попытку добавляется еще одна часть туловища висельника (обычно их 6: голова, туловище, 2 руки и 2 ноги.

```python
import random

dic = ['год', 'человек', 'время', 'дело', 'жизнь', 'день', 'рука', 'раз', 'работа', 'слово', 'место', 'лицо', 'друг',
       'глаз','вопрос', 'дом', 'сторона', 'страна', 'мир', 'случай', 'голова', 'ребенок', 'сила']


# функция для загадывания слова из существующего списка
def get_word():
    word = [x.upper() for x in random.choice(dic)]
    return word
    # print(word)


# Показываем количество оставшихся попыток
def display_hangman(tries):
    if tries!=0:
        return print(f'{tries} tries left!')
    else:
       return print('You Loose!')


# Основная игровая функция
def play(word):
    word_completion = ['_' for x in range(len(word))]  # строка, содержащая символы _ на каждую букву задуманного слова
    guessed = False  # сигнальная метка
    guessed_letters = []  # список уже названных букв
    guessed_words = []  # список уже названных слов
    tries = 6  # количество попыток

    again='д'
    print('Давай играть в угадайку слов!')
    while again=='д' or 'да':
        print(*word_completion)
        display_hangman(tries)
        user_input = list(input().upper())
        if len(user_input) != len(word):
            print('Неверная длина слова! Попробуй снова.')
            continue
        for i in user_input:
            if i in '!@#$%^&*()-_=+0123456789<>?":;[]{}':
                print('Вы использовали недопустимые символы! введите слово заново!')
                break
        if user_input in guessed_letters and user_input != word:
            print("Вы уже вводили эту букву!")
            continue
        for j in range(len(user_input)):
            # print(user_input[j], word[j])
            if user_input[j] == word[j]:
                guessed_letters.append(word[j])
                word_completion[j] = user_input[j]
        if user_input in guessed_words:
            print("Вы уже вводили это слово! Попробуйте снова")
            continue

        tries -= 1
        print(*word_completion)

        if user_input == word:
            print('Поздравляю, Вы победили!')
            guessed = True
        else:
            display_hangman(tries)

        guessed_words.append(user_input)

        if tries == 0:
            print()
            print('Вот загаданное слово:', *word)
            print('Вы проиграли!')
        if guessed == True or tries == 0:
            print('Желаете начать новую игру?')
            again = input('Введите "Д" если желаете сыграть снова, либо "Н" если хотите закончить.')


play(get_word())
```
