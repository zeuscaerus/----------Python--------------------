–ü—Ä–∏ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏ –Ω–∞ —Å–∞–π—Ç–∞—Ö —Ç—Ä–µ–±—É–µ—Ç—Å—è –≤–≤–æ–¥–∏—Ç—å –ø–∞—Ä–æ–ª—å –¥–≤–∞–∂–¥—ã. –≠—Ç–æ —Å–¥–µ–ª–∞–Ω–æ –¥–ª—è –±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç–∏, –ø–æ—Å–∫–æ–ª—å–∫—É —Ç–∞–∫–æ–π –ø–æ–¥—Ö–æ–¥ —É–º–µ–Ω—å—à–∞–µ—Ç –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –Ω–µ–≤–µ—Ä–Ω–æ–≥–æ –≤–≤–æ–¥–∞ –ø–∞—Ä–æ–ª—è.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—Ä–∞–≤–Ω–∏–≤–∞–µ—Ç –ø–∞—Ä–æ–ª—å –∏ –µ–≥–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ. –ï—Å–ª–∏ –æ–Ω–∏ —Å–æ–≤–ø–∞–¥–∞—é—Ç, —Ç–æ –ø—Ä–æ–≥—Ä–∞–º–º–∞ –≤—ã–≤–æ–¥–∏—Ç: ¬´–ü–∞—Ä–æ–ª—å –ø—Ä–∏–Ω—è—Ç¬ª, –∏–Ω–∞—á–µ: ¬´–ü–∞—Ä–æ–ª—å –Ω–µ –ø—Ä–∏–Ω—è—Ç¬ª.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è –¥–≤–µ —Å—Ç—Ä–æ–∫–∏.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

```python
a,b=input(), input()
if a==b:
    print('–ü–∞—Ä–æ–ª—å –ø—Ä–∏–Ω—è—Ç')
else:
    print('–ü–∞—Ä–æ–ª—å –Ω–µ –ø—Ä–∏–Ω—è—Ç')
```

–ß–µ—Ç–Ω–æ–µ –∏–ª–∏ –Ω–µ—á–µ—Ç–Ω–æ–µ?

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, —è–≤–ª—è–µ—Ç—Å—è —á–∏—Å–ª–æ —á–µ—Ç–Ω—ã–º –∏–ª–∏ –Ω–µ—á–µ—Ç–Ω—ã–º.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ ¬´–ß–µ—Ç–Ω–æ–µ¬ª, –µ—Å–ª–∏ —á–∏—Å–ª–æ —á–µ—Ç–Ω–æ–µ, –∏ ¬´–ù–µ—á–µ—Ç–Ω–æ–µ¬ª ‚Äî –µ—Å–ª–∏ —á–∏—Å–ª–æ –Ω–µ—á–µ—Ç–Ω–æ–µ.

```python
i=int(input())
if i % 2 != 0:
    print('–ù–µ—á–µ—Ç–Ω–æ–µ')
else:
    print('–ß–µ—Ç–Ω–æ–µ')
```

–°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ 

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–æ–≤–µ—Ä—è–µ—Ç, —á—Ç–æ –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–µ—Ç—ã—Ä–µ—Ö–∑–Ω–∞—á–Ω–æ–≥–æ —á–∏—Å–ª–∞ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–µ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ: —Å—É–º–º–∞ –ø–µ—Ä–≤–æ–π –∏ –ø–æ—Å–ª–µ–¥–Ω–µ–π —Ü–∏—Ñ—Ä —Ä–∞–≤–Ω–∞ —Ä–∞–∑–Ω–æ—Å—Ç–∏ –≤—Ç–æ—Ä–æ–π –∏ —Ç—Ä–µ—Ç—å–µ–π —Ü–∏—Ñ—Ä.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –æ–¥–Ω–æ —Ü–µ–ª–æ–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–µ—Ç—ã—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ ¬´–î–ê¬ª, –µ—Å–ª–∏ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è, –∏ ¬´–ù–ï–¢¬ª ‚Äî –µ—Å–ª–∏ –Ω–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è.

```python
a,b,c,d=map(int,input())
if a+d == b-c:
    print('–î–ê')
else:
    print('–ù–ï–¢')
```

–†–æ—Å–∫–æ–º–Ω–∞–¥–∑–æ—Ä

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, —Ä–∞–∑—Ä–µ—à–µ–Ω –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é –¥–æ—Å—Ç—É–ø –∫ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-—Ä–µ—Å—É—Ä—Å—É –∏–ª–∏ –Ω–µ—Ç.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äî –≤–æ–∑—Ä–∞—Å—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç ¬´–î–æ—Å—Ç—É–ø —Ä–∞–∑—Ä–µ—à–µ–Ω¬ª –µ—Å–ª–∏ –≤–æ–∑—Ä–∞—Å—Ç –Ω–µ –º–µ–Ω–µ–µ 18, –∏ ¬´–î–æ—Å—Ç—É–ø –∑–∞–ø—Ä–µ—â–µ–Ω¬ª –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

```python
x=int(input())
if x >= 18:
    print('–î–æ—Å—Ç—É–ø —Ä–∞–∑—Ä–µ—à–µ–Ω')
else:
    print('–î–æ—Å—Ç—É–ø –∑–∞–ø—Ä–µ—â–µ–Ω')
```

–ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∞—è –ø—Ä–æ–≥—Ä–µ—Å—Å–∏—è

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, —è–≤–ª—è—é—Ç—Å—è –ª–∏ —Ç—Ä–∏ –∑–∞–¥–∞–Ω–Ω—ã—Ö —á–∏—Å–ª–∞ (–≤ —É–∫–∞–∑–∞–Ω–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ) –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω—ã–º–∏ —á–ª–µ–Ω–∞–º–∏ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–π –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–∏.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è —Ç—Ä–∏ —á–∏—Å–ª–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ ¬´YES¬ª –∏–ª–∏ ¬´NO¬ª (–±–µ–∑ –∫–∞–≤—ã—á–µ–∫) –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏. 

```python
a, b, c = int(input()), int(input()), int(input())
if a+a == b and b+a ==c or a-b == b and b-b ==c:
    print('YES')
else:
    print('NO')
```

–ù–∞–∏–º–µ–Ω—å—à–µ–µ –∏–∑ –¥–≤—É—Ö —á–∏—Å–µ–ª

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –∏–∑ –¥–≤—É—Ö —á–∏—Å–µ–ª.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –¥–≤–∞ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –∏–∑ –¥–≤—É—Ö —á–∏—Å–µ–ª.

```python
a,b = int(input()), int(input())
if a>b:
    print(b)
else:
    print(a)
```

–ù–∞–∏–º–µ–Ω—å—à–µ–µ –∏–∑ —á–µ—Ç—ã—Ä—ë—Ö —á–∏—Å–µ–ª üå∂Ô∏è

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –∏–∑ —á–µ—Ç—ã—Ä—ë—Ö —á–∏—Å–µ–ª.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —á–µ—Ç—ã—Ä–µ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –∏–∑ —á–µ—Ç—ã—Ä—ë—Ö —á–∏—Å–µ–ª.

```python
a,b,c,d = int(input()), int(input()), int(input()),int(input())
if a<b<c<d:
    print (a)
else:
    if b<c<d:
        print(b)
    else:
        if c<d:
            print(c)
        else:
            print(d)
```

–í–æ–∑—Ä–∞—Å—Ç–Ω–∞—è –≥—Ä—É–ø–ø–∞

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ –≤–≤–µ–¥—ë–Ω–Ω–æ–º—É –≤–æ–∑—Ä–∞—Å—Ç—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è —Å–æ–æ–±—â–∞–µ—Ç, –∫ –∫–∞–∫–æ–π –≤–æ–∑—Ä–∞—Å—Ç–Ω–æ–π –≥—Ä—É–ø–ø–µ –æ–Ω –æ—Ç–Ω–æ—Å–∏—Ç—Å—è:

–¥–æ 13 –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ ‚Äì –¥–µ—Ç—Å—Ç–≤–æ;
–æ—Ç 14 –¥–æ 24 ‚Äì –º–æ–ª–æ–¥–æ—Å—Ç—å;
–æ—Ç 25 –¥–æ 59 ‚Äì –∑—Ä–µ–ª–æ—Å—Ç—å;
–æ—Ç 60 ‚Äì —Å—Ç–∞—Ä–æ—Å—Ç—å.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äì –≤–æ–∑—Ä–∞—Å—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –Ω–∞–∑–≤–∞–Ω–∏–µ –≤–æ–∑—Ä–∞—Å—Ç–Ω–æ–π –≥—Ä—É–ø–ø—ã.

```python
a = int(input())
if a>=60:
    print('—Å—Ç–∞—Ä–æ—Å—Ç—å')
else:
    if 24<a<60:
        print('–∑—Ä–µ–ª–æ—Å—Ç—å')
    else:
        if 13<a<26:
            print('–º–æ–ª–æ–¥–æ—Å—Ç—å')
        else:
            print('–¥–µ—Ç—Å—Ç–≤–æ')
```

–¢–æ–ª—å–∫–æ + üå∂Ô∏è

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç—ã–≤–∞–µ—Ç —Ç—Ä–∏ —á–∏—Å–ª–∞ –∏ –ø–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ—Ç —Å—É–º–º—É —Ç–æ–ª—å–∫–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è —Ç—Ä–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äì —Å—É–º–º—É –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ï—Å–ª–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª –Ω–µ—Ç, —Ç–æ —Å–ª–µ–¥—É–µ—Ç –≤—ã–≤–µ—Å—Ç–∏ 0.

```python
l2=[]
a,b,c = int(input()), int(input()), int(input())
l=[a,b,c]
for i in l:
    if i>=0:
        l2.append(i)
print(sum(l2))
```

4.2 –õ–æ–≥–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ x.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

```python
x=int(input())
if x > -1 and x < 17:
    print('–ü—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç')
else:
    print('–ù–µ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç')
```

–ü—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç—å 2

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ x –∏ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –ª–∏ –¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ —É–∫–∞–∑–∞–Ω–Ω—ã–º –ø—Ä–æ–º–µ–∂—É—Ç–∫–∞–º.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ x.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

```python
x=int(input())
if -3<x<7:
    print('–ù–µ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç')
else:
    print('–ü—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç')
```

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ x –∏ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –ª–∏ –¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ —É–∫–∞–∑–∞–Ω–Ω—ã–º –ø—Ä–æ–º–µ–∂—É—Ç–∫–∞–º.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ x.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

```python
x=int(input())
if -30<x<=-2 or 7<x<=25:
    print('–ü—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç')
else:
    print('–ù–µ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç')
```

–ö—Ä–∞—Å–∏–≤–æ–µ —á–∏—Å–ª–æ üå∂Ô∏è

–ù–∞–∑–æ–≤–µ–º —á–∏—Å–ª–æ –∫—Ä–∞—Å–∏–≤—ã–º, –µ—Å–ª–∏ –æ–Ω–æ —è–≤–ª—è–µ—Ç—Å—è —á–µ—Ç—ã—Ä–µ—Ö–∑–Ω–∞—á–Ω—ã–º –∏ –¥–µ–ª–∏—Ç—Å—è –Ω–∞—Ü–µ–ª–æ –Ω–∞ 7 –∏–ª–∏ –Ω–∞ 17. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –æ–ø—Ä–µ–¥–µ–ª—è—é—â—É—é, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –≤–≤–µ–¥—ë–Ω–Ω–æ–µ —á–∏—Å–ª–æ –∫—Ä–∞—Å–∏–≤—ã–º. –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ ¬´YES¬ª, –µ—Å–ª–∏ —á–∏—Å–ª–æ —è–≤–ª—è–µ—Ç—Å—è –∫—Ä–∞—Å–∏–≤—ã–º, –∏–ª–∏ ¬´NO¬ª –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

```python
x=int(input())
if 999<x<10000:
    if (x%7)==0 or (x%17)==0:
        print('YES')
    else:
        print('NO')
else:
    print('NO')
```

–ù–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Ç—Ä–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–ª–∞ –∏ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –ª–∏ –Ω–µ–≤—ã—Ä–æ–∂–¥–µ–Ω–Ω—ã–π —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ —Å —Ç–∞–∫–∏–º–∏ —Å—Ç–æ—Ä–æ–Ω–∞–º–∏.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ç—Ä–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ ¬´YES¬ª –∏–ª–∏ ¬´NO¬ª –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç, –µ—Å–ª–∏ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞.

```python
a,b,c = int(input()), int(input()), int(input())
if  a+b>c and a+c>b and b+c>a:
    print('YES')
else:
    print('NO')
```

–í–∏—Å–æ–∫–æ—Å–Ω—ã–π –≥–æ–¥

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –≥–æ–¥ —Å –¥–∞–Ω–Ω—ã–º –Ω–æ–º–µ—Ä–æ–º –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º. –ï—Å–ª–∏ –≥–æ–¥ —è–≤–ª—è–µ—Ç—Å—è –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º, —Ç–æ –≤—ã–≤–µ–¥–∏—Ç–µ ¬´YES¬ª, –∏–Ω–∞—á–µ –≤—ã–≤–µ–¥–∏—Ç–µ ¬´NO¬ª.

–ì–æ–¥ —è–≤–ª—è–µ—Ç—Å—è –≤–∏—Å–æ–∫–æ—Å–Ω—ã–º, –µ—Å–ª–∏ –µ–≥–æ –Ω–æ–º–µ—Ä –∫—Ä–∞—Ç–µ–Ω 4, –Ω–æ –Ω–µ –∫—Ä–∞—Ç–µ–Ω 100, –∏–ª–∏ –µ—Å–ª–∏ –æ–Ω –∫—Ä–∞—Ç–µ–Ω 400.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

```python
y = int(input())
if y%100!=0 and y%4==0 or y%400==0:
    print('YES')
else:
    print('NO')
```

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —á–µ—Ç—ã—Ä–µ —á–∏—Å–ª–∞ –æ—Ç 1 –¥–æ 8.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –®–∞—Ö–º–∞—Ç–Ω–∞—è –ª–∞–¥—å—è —Ö–æ–¥–∏—Ç –ø–æ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª–∏ –∏–ª–∏ –≤–µ—Ä—Ç–∏–∫–∞–ª–∏.

```python
x1,y1,x2,y2 = int(input()), int(input()), int(input()), int(input())
if x1==x2 or y1==y2:
    print('YES')
else:
    print('NO')
```

–•–æ–¥ –∫–æ—Ä–æ–ª—è üå∂Ô∏è

–î–∞–Ω—ã –¥–≤–µ —Ä–∞–∑–ª–∏—á–Ω—ã–µ –∫–ª–µ—Ç–∫–∏ —à–∞—Ö–º–∞—Ç–Ω–æ–π –¥–æ—Å–∫–∏. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É,  –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, –º–æ–∂–µ—Ç –ª–∏ –∫–æ—Ä–æ–ª—å –ø–æ–ø–∞—Å—Ç—å —Å –ø–µ—Ä–≤–æ–π –∫–ª–µ—Ç–∫–∏ –Ω–∞ –≤—Ç–æ—Ä—É—é –æ–¥–Ω–∏–º —Ö–æ–¥–æ–º. –ü—Ä–æ–≥—Ä–∞–º–º–∞ –ø–æ–ª—É—á–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ —á–µ—Ç—ã—Ä–µ —á–∏—Å–ª–∞ –æ—Ç 1 –¥–æ 8 –∫–∞–∂–¥–æ–µ, –∑–∞–¥–∞—é—â–∏–µ –Ω–æ–º–µ—Ä —Å—Ç–æ–ª–±—Ü–∞ –∏ –Ω–æ–º–µ—Ä —Å—Ç—Ä–æ–∫–∏ —Å–Ω–∞—á–∞–ª–∞ –¥–ª—è –ø–µ—Ä–≤–æ–π –∫–ª–µ—Ç–∫–∏, –ø–æ—Ç–æ–º –¥–ª—è –≤—Ç–æ—Ä–æ–π –∫–ª–µ—Ç–∫–∏. –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ ¬´YES¬ª, –µ—Å–ª–∏ –∏–∑ –ø–µ—Ä–≤–æ–π –∫–ª–µ—Ç–∫–∏ —Ö–æ–¥–æ–º –∫–æ—Ä–æ–ª—è –º–æ–∂–Ω–æ –ø–æ–ø–∞—Å—Ç—å –≤–æ –≤—Ç–æ—Ä—É—é, –∏–ª–∏ ¬´NO¬ª –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —á–µ—Ç—ã—Ä–µ —á–∏—Å–ª–∞ –æ—Ç 1 –¥–æ 8.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –®–∞—Ö–º–∞—Ç–Ω—ã–π –∫–æ—Ä–æ–ª—å —Ö–æ–¥–∏—Ç –ø–æ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª–∏, –≤–µ—Ä—Ç–∏–∫–∞–ª–∏ –∏ –¥–∏–∞–≥–æ–Ω–∞–ª–∏, –Ω–æ —Ç–æ–ª—å–∫–æ –Ω–∞ 1 –∫–ª–µ—Ç–∫—É.

```python
x1,y1,x2,y2 = int(input()), int(input()), int(input()), int(input())
if x1-2<x2<x1+2 and y1-2<y2<y1+2: 
    print('YES')
else:
    print('NO')
```

–ì–æ–Ω–∫–∞ —Å–ø–∏–¥—Å—Ç–µ—Ä–æ–≤

–ó—É–º –±—Ä–æ—Å–∏–ª –≤—ã–∑–æ–≤ –§–ª—ç—à—É –∏ –ø—Ä–µ–¥–ª–æ–∂–∏–ª –µ–º—É —á–µ—Å—Ç–Ω—ã–π –ø–æ–µ–¥–∏–Ω–æ–∫ –≤ –≤–∏–¥–µ –≥–æ–Ω–∫–∏ –≤–æ–∫—Ä—É–≥ –º–∞–≥–Ω–µ—Ç–∞—Ä–∞. –í —Å–ª—É—á–∞–µ –ø—Ä–æ–∏–≥—Ä—ã—à–∞ —ç—Ç–∞ –Ω–µ–π—Ç—Ä–æ–Ω–Ω–∞—è –∑–≤–µ–∑–¥–∞ –∑–∞—Ä—è–¥–∏—Ç—Å—è –∏ —É–Ω–∏—á—Ç–æ–∂–∏—Ç –º–∏—Ä, –ø–æ—ç—Ç–æ–º—É –§–ª—ç—à —Ä–µ—à–∏–ª –Ω–µ —Ä–∏—Å–∫–æ–≤–∞—Ç—å –±–µ–∑ –ø—Ä–∏—á–∏–Ω—ã, –∏ —É–∑–Ω–∞—Ç—å —É —Å–≤–æ–µ–≥–æ –¥—Ä—É–≥–∞ –¶–∏—Å–∫–æ –†–∞–º–æ–Ω–∞ –µ—Å—Ç—å –ª–∏ —Å–º—ã—Å–ª –ø—Ä–∏–Ω–∏–º–∞—Ç—å –≤—ã–∑–æ–≤. –¶–∏—Å–∫–æ –ø–æ–ª—É—á–∏–ª –¥–∞–Ω–Ω—ã–µ, —á—Ç–æ —Å–∫–æ—Ä–æ—Å—Ç—å –ó—É–º–∞ —Ä–∞–≤–Ω–∞ n, –∞ —Å–∫–æ—Ä–æ—Å—Ç—å –§–ª—ç—à–∞ —Ä–∞–≤–Ω–∞ k.

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ—Ç–≤–µ—Ç –¶–∏—Å–∫–æ –Ω–∞ –≤–æ–ø—Ä–æ—Å –§–ª—ç—à–∞.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ n –∏ k, —Å–∫–æ—Ä–æ—Å—Ç—å –ó—É–º–∞ –∏ –§–ª—ç—à–∞.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ï—Å–ª–∏ –ó—É–º –±—ã—Å—Ç—Ä–µ–µ –§–ª—ç—à–∞ –Ω—É–∂–Ω–æ –≤—ã–≤–µ—Å—Ç–∏ ¬´NO¬ª, –µ—Å–ª–∏ –§–ª—ç—à –±—ã—Å—Ç—Ä–µ–µ –ó—É–º–∞ –Ω—É–∂–Ω–æ –≤—ã–≤–µ—Å—Ç–∏ ¬´YES¬ª, –µ—Å–ª–∏ –∏—Ö —Å–∫–æ—Ä–æ—Å—Ç–∏ —Ä–∞–≤–Ω—ã –Ω—É–∂–Ω–æ –≤—ã–≤–µ—Å—Ç–∏ "Don't know".

```python
a, b, c = int(input()), int(input()), int(input())
if a==b==c:
    print('–†–∞–≤–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π')
else:
    if a!=b and c!=b and a!=c:
        print('–†–∞–∑–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π')
    else:
        print('–†–∞–≤–Ω–æ–±–µ–¥—Ä–µ–Ω–Ω—ã–π')
```

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:

```python
num=int(input())
x=num%2
if 0<=num<37:
    if num==0:
        print('–∑–µ–ª–µ–Ω—ã–π')
    elif 0<num<11 or 18<num<29:
        if x==0:
            print('—á–µ—Ä–Ω—ã–π')
        else:
            print('–∫—Ä–∞—Å–Ω—ã–π')
    elif 10<num<19 or 28<num<37:
        if x==0:
            print('–∫—Ä–∞—Å–Ω—ã–π')
        else:
            print('—á–µ—Ä–Ω—ã–π')
else:
    print('–æ—à–∏–±–∫–∞ –≤–≤–æ–¥–∞')
```
